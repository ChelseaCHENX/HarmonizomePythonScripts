# Map Sample Metadata to Sample ID

```python
lst = []

for i,col in enumerate(matrix.columns):
    col_num = i+1
    if donor_meta.loc[sample_meta.loc[int(col), 'donor_id'], 'ever_tbi_w_loc'] == 'N':
        sample = (
            str(sample_meta.loc[int(col), 'structure_name'])
            +' ,'+
            str(sample_meta.loc[int(col), 'hemisphere'])
            +'_'+
            str(donor_meta.loc[sample_meta.loc[int(col), 'donor_id'], 'act_demented'])
    #     )
            +'_'+
            'No TBI'
                 )
    else:
        sample = (
            str(sample_meta.loc[int(col), 'structure_name'])
            +' ,'+
            str(sample_meta.loc[int(col), 'hemisphere'])
            +'_'+
            str(donor_meta.loc[sample_meta.loc[int(col), 'donor_id'], 'act_demented'])
            +'_'+
            'TBI'
                 )
    lst.append(sample)
    
matrix.columns = lst
```

```python
normalized_matrix = uf.removeAndImpute(normalized_matrix)
```

```python
normalized_matrix.shape
```

# Log2 Transform

```python
normalized_matrix = uf.log2(normalized_matrix)
```

```python
normalized_matrix.head()
```

# Normalize Matrix (Quantile Normalize the matrix for the columns)

```python
normalized_matrix = uf.quantileNormalize(normalized_matrix)
```

```python
normalized_matrix.head()
```

# Normalize Matrix (z-score the rows)

```python
uf.zscore(normalized_matrix, 'row')
```

```python
normalized_matrix.head()
```

# Merge Like Column (by taking the mean)

```python
normalized_matrix = uf.merge(normalized_matrix, 'column', 'mean')
```

```python
normalized_matrix.shape
```

# Map Gene Symbols To Up-to-date Approved Gene Symbols

```python
uf.mapgenesymbols(normalized_matrix)
```

```python
normalized_matrix.shape
```

```python
normalized_matrix = uf.merge(normalized_matrix, 'row', 'mean')
```

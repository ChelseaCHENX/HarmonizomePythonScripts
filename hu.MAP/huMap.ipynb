```python
hu_ppi_df[0] = lst
```

```python
lst = []

for index in hu_ppi_df.index:
    if 'E' in hu_ppi_df.ix[index, 1]:
        lst.append(np.nan)
    elif float(hu_ppi_df.ix[index, 1]) in mapping_df.index:
        lst.append(mapping_df.ix[float(hu_ppi_df.ix[index, 1]), 'Approved Symbol'])
    else:
        lst.append(np.nan)
```
